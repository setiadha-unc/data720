## Word frequency analysis
* Counting the frequency of words in a text file is an important metric in text mining.
* This notebook is a straightforward implementation of a way to count the number of times a word occurs in a text file.



### The name of the file is assumed to be stored in `file_name`.

```python
file_name = "Unit-2-8-Sample-Text.txt"
```

### Suggested algorithm
* Use a dictionary to count the word frequency
* Start with an empty dictionary
* Open the file and read line by line.
* For each line:
    * use strip to get the words into a list (assume word separator is white space)
    * for each word in the list:
        * if the word exists in the dictionary:
            * increase its count by 1
        * else add the word as a new key to the dictionary with a count of 1
        
* Use the dictionary to print out the word frequency

```python
word_dict = {}
with open(file_name) as f:
    for line in f:
        word_list = line.split()
        for word in word_list:
            print(word)
            if (word in word_dict):
                word_dict[word] += 1
            else:
                word_dict[word] = 1
print(f"Number of unique words = {len(word_dict)}")               
print(word_dict)
                
```

### Hmm, seems to work, but some punctuation marks are getting included as words.
### Let us get rid of those using  `strip`

```python
word_dict = {}
with open(file_name) as f:
    for line in f:
        word_list = line.split()
        for word in word_list:
            print(word, end = ' ')
            word_clean =  word.strip(',"?.\'')
            print(word_clean)
            if (word_clean in word_dict):
                word_dict[word_clean] += 1
            else:
                word_dict[word_clean] = 1

print(f"Number of unique words = {len(word_dict)}")                
print(word_dict)
```
