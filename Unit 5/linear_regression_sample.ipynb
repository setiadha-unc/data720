```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Sample Data: Square Footage vs. House Price
square_footage = np.array([750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200]).reshape(-1, 1)
house_price = np.array([150000, 160000, 170000, 180000, 190000, 200000, 210000, 220000, 230000, 240000])
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(square_footage, house_price, test_size=0.2, random_state=42)
```

```python
# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict house prices
y_pred = model.predict(X_test)
y_pred
```

```python
# Plot results
plt.scatter(X_test, y_test, color='blue', label="Actual Prices")
plt.plot(X_test, y_pred, color='red', linewidth=2, label="Predicted Line")
plt.xlabel("Square Footage")
plt.ylabel("House Price ($)")
plt.title("Linear Regression: Square Footage vs. House Price")
plt.legend()
plt.show()
```

```python
# Print model coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```
