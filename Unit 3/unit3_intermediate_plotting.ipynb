```python
import numpy as np
import matplotlib.pyplot as plt
import plot_variables as pv
```

```python
# LINSPACE, CREATE LINEAR 100 DATA BETWEEN 1 AND 100
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)


fig, axes = plt.subplots(nrows=3, ncols=1)
axes[0].plot(x, y_sin)
axes[0].set_title("Sin(x)")

axes[1].plot(x, y_cos)
axes[1].set_title("Cos(x)")


axes[2].plot(x, y_tan)
axes[2].set_title("Tan(x)")

plt.show()
```

```python
high_index = [j for j, total in enumerate(pv.total) if total > 200]
x = [pv.country[i] for i in high_index]
y = [pv.total[i] for i in high_index]

plt.figure(figsize=(8,6))
plt.bar(x, y)
plt.xlabel("Counrty")
plt.ylabel("Total CO2 emissions")
# to rotate the x label
plt.xticks(rotation=90)
plt.show()
```

```python
low_index = [i for i, total in enumerate(pv.total) if total < 1]
x_low = [pv.country[i] for i in low_index]
y_low = [pv.total[i] for i in low_index]

plt.figure(figsize=(8,6))
plt.bar(x_low, y_low)
plt.xlabel("Counrty")
plt.ylabel("Total CO2 emissions")
plt.xticks(rotation=90)
plt.show()
```

```python
colors = []
for total in y:
    if total > 1000:
        colors.append("red")
    else:
        colors.append("blue")
```

```python
plt.figure(figsize=(8,6))
plt.bar(x, y, color=colors)
plt.xlabel("Counrty")
plt.ylabel("Total CO2 emissions")
plt.xticks(rotation=90)
plt.show()
```

```python
label_dict = {
    "USA": 0,
    "China": 0,
    "India": 0,
    "Everyone else": 0
}

for country, total in zip(pv.country, pv.total):
    if country in list(label_dict.keys()):
        label_dict[country] += total
    else:
        label_dict["Everyone else"] += total

slices = list(label_dict.values())
labels = list(label_dict.keys())

plt.pie(slices, labels=labels, colors=['red', 'blue', 'green', 'purple'])
plt.show()
```
