```python
x = pv.internet_year
y1 = pv.internet_high_income
y2 = pv.internet_middle_income
y3 = pv.internet_low_income

plt.figure(figsize=(8,6))
plt.plot(x, y1, marker="o", linestyle="--")
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("Year")
plt.ylabel("Internet users (%)")

plt.show()
```

```python
x = pv.pops
y = pv.total

plt.figure(figsize=(8,6))
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.xlabel("Population")
plt.ylabel("CO2 emissions")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x, y, s=pv.coal, c=pv.per_capita, cmap="viridis")
plt.xlabel("Population")
plt.ylabel("CO2 emissions")
plt.colorbar(label="Per capita emissions")
plt.xscale("log")
plt.yscale("log")
plt.show()
```

```python
plt.figure(figsize=(8,6))
# S = size, C Color scale
plt.scatter(x, y, s=pv.coal, c=pv.per_capita, cmap="viridis", alpha=0.6)
plt.xlabel("Population")
plt.ylabel("CO2 emissions")
plt.colorbar(label="Per capita emissions")
plt.xscale("log")
plt.yscale("log")
plt.xlim([10**4, 10**10])
plt.ylim([10**-1, 10**5])
plt.show()
```
