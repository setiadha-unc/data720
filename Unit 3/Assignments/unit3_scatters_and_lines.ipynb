The following cell provides some data.

```python
import random

num_points = 10
x1 = range(num_points)
y1 = [random.randint(0, 10) for _ in range(num_points)]
z1 = [random.randint(5, 20) for _ in range(num_points)]

x2 = range(num_points)
y2 = [random.randint(0, 10) for _ in range(num_points)]
z2 = [random.randint(5, 20) for _ in range(num_points)]
```

In the cell below, correctly import Matplotlib

```python
import matplotlib.pyplot as plt
```

Create a new figure with one subplot

```python
fig, ax = plt.subplots()
```

Plot the `x1` and `y1` variables as a line plot with their data on their corresponding axes.

```python
fig, ax = plt.subplots()
ax.plot(x1, y1)
plt.show()
```

Add the plot of `x2` and `y2` to the same plot.

```python
fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.plot(x2, y2)
plt.show()
```

Change the colors of the lines to `lightblue` (`x1/y1`) and `darkblue` (`x2/y2`).

```python
fig, ax = plt.subplots()
ax.plot(x1, y1, color="lightblue")
ax.plot(x2, y2, color="darkblue")
plt.show()
```

Label the x-axis `X` and y-axis `Y`

```python
fig, ax = plt.subplots()
ax.plot(x1, y1, color="lightblue")
ax.plot(x2, y2, color="darkblue")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

Title the plot `Random data`

```python
fig, ax = plt.subplots()
ax.plot(x1, y1, color="lightblue")
ax.plot(x2, y2, color="darkblue")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Random Data")
plt.show()
```

Now switch the plot to a scatter plot.

```python
fig, ax = plt.subplots()
ax.scatter(x1, y1, color="lightblue")
ax.scatter(x2, y2, color="darkblue")
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

Finally, make the marker sizes for `x1/y1` correspond to `z1` and `x2/y2` correspond to `z2`

```python
fig, ax = plt.subplots()
ax.scatter(x1, y1, color="lightblue", s=z1)
ax.scatter(x2, y2, color="darkblue", s=z2)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```
