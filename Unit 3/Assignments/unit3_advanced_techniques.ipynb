The following cell provides some data.

```python
import random

num_points = 10

# line
x1 = range(num_points)
y1 = [random.randint(0, 10) for _ in range(num_points)]

# scatter
x2 = [random.randint(0, 10) + (random.randint(0, 10) * 0.1)  for _ in range(num_points**2)]
y2 = [random.randint(0, 10) + (random.randint(0, 10) * 0.1) for _ in range(num_points**2)]
z2 = [random.randint(5, 20) for _ in range(num_points**2)]

# pie
pie = 20
cake = 58
cookies = 22
```

In the cell below, correctly import Matplotlib

```python
import matplotlib.pyplot as plt
```

Create a line plot with `x1` and `y1` on their respective axes. Label the x-axis `X` and y-axis `Y`.

```python
fig, ax1 = plt.subplots()

ax1.plot(x1, y1)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Create a scatter plot with `x2` and `y2` on their respective axes, with the size of the points set by `z2` and a transparency of 50%. Label the x-axis `X` and y-axis `Y`. 

```python
fig, ax2 = plt.subplots()

ax2.scatter(x2, y2, s=z2, alpha=.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

Make a pie chart of the three categories labeled `# pie`. 

```python
plt.pie([pie, cake, cookies], labels=['pie', 'cake', 'cookies'])
plt.title('PIE')
plt.show()
```

Create subplots using a `GridSpec` that will produce the output shown (the import statement has been provided). Make sure to set the title of each subplot.


```python
# IMPORT THE LIB
from matplotlib.gridspec import GridSpec
```

```python
# SET THE BASE
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 2, figure=fig)
ax1 = fig.add_subplot(gs[0,0])
ax1.plot(x1, y1)
ax1.set_title("Line plot", fontsize=18)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")

ax2 = fig.add_subplot(gs[0,1])
ax2.pie([pie, cake, cookies], labels=['pie', 'cake', 'cookies'])
ax2.set_title("Pie Chart", fontsize=18)

ax3 = fig.add_subplot(gs[1, :])
ax3.scatter(x2, y2, s=z2, alpha=.5)
ax3.set_title("Scatter plot", fontsize=18)
ax3.set_xlabel("X")
ax3.set_ylabel("Y")

plt.tight_layout()
plt.show()
```
