```python
import pandas as pd
```

# Loading data

```python
# Read csv
data = pd.read_csv("usda_farm_indices.csv")
data.head()
```

```python
# Create pivot table (columns from values)
data = data.pivot(index="Year", columns="Attribute", values="Value")
totals = [col for col in data.columns if "Total" in col]
data = data[totals]
data.head()
```

# Sampling data

```python
# Random sample
df1 = data.sample(5, random_state=1)
df1.head()
```

```python
# Different random sample using different random state
df2 = data.sample(5, random_state=2)
df2.head()
```

```python
# Concatenate two dfs
concat = pd.concat([df1, df2], axis=0)
concat
```

```python
# Create two dfs from above, but with different columns
df4 = concat.drop(columns=concat.columns[:4])
df5 = concat.drop(columns=concat.columns[4:])
# Dsiplay first
df4.head()
```
